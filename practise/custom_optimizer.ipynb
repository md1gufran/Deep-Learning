```python
import tensorflow as tf
from tensorflow.keras.optimizers import Optimizer
from tensorflow import keras
from tensorflow.keras.optimizers import Adagrad, Optimizer
```

```python
class CustomOptimizer(Optimizer):
    def __init__(self, learning_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8, decay=0.0, **kwargs):
        super(CustomOptimizer, self).__init__(**kwargs)
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.decay = decay

    def get_updates(self, params, grads, states):
        updates = []
        for p, g, m in zip(params, grads, states):
            # Get the accumulated history for Adagrad
            history = m.get("history", tf.zeros_like(p))

            # Update history with current gradient square
            history = tf.assign(history, history + tf.square(g))

            # Adagrad update with decaying learning rate
            adagrad_update = g / (tf.sqrt(history) + self.epsilon) * self.learning_rate * (1.0 - self.decay)

            # BGD update
            bgd_update = - self.learning_rate * g

            # Combine Adagrad and BGD updates
            combined_update = adagrad_update + bgd_update

            # Update parameters
            new_p = tf.assign(p, p + combined_update)

            # Update state for Adagrad
            updates.append(tf.assign(m, {"history": history}))
            updates.append(new_p)
        return updates
```

```python
x = df.drop("salary",axis="columns")
y = df["salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = keras.Sequential([
    keras.layers.Dense(2*df.shape[1],input_shape=(df.shape[1],),activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10,activation="relu"),
    keras.layers.Dense(1,activation="linear"),
])
model.compile(
    optimizer = CustomOptimizer(learning_rate=0.01),
    loss = "mae",
    metrics = ["accuracy"]
)
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50)
```
